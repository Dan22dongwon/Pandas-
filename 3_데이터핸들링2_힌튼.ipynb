# pandas란? 

* 데이터를 효과적으로 처리하고, 보여줄 수 있도록 도와주는 라이브러리입니다.
* Numpy와 함께 사용되어 다양한 연계적인 기능을 제공합니다.
* 인덱스(Index)에 따라 데이터를 나열하므로 사전(Dictionary) 자료형에 가깝습니다.
* 시리즈(Series)를 기본적인 자료형으로 사용합니다.
* 엑셀(Excel)과 거의 유사합니다

## Pandas 진짜 조금 더 알아보기
* 분석을 할 때는 열(column) 단위로

* Series는 **1차원 배열**의 형태를 가짐, 인덱스라는 한 가지 기준에 의하여 데이터 저장
, 시리즈는 단일한 형태의 객체를 가짐
→ (일차원 ndarray로서 축 레이블을 가지고 있고, 배열 형태, dict, 스칼라 값 등을 저장 할 수 있어 dictionary는 키:값 쌍을 저장하는 파이썬의 기본 자료구조로서, 내장 메소드를 통해데이터를 조작할 수 있음)

* Series는 dictionary와 달리 정렬되어 있고, 복잡한 데이터 조작이나 파일 입출력을 할 때 유용한 메소드와 속성을 가지고 있다.

* Series를 쓰면 dictionary보다 더 효율적이고 간편하게 데이터를 다룰 수 있음

* DataFrame은 **2차원 배열**의 형태를 가짐, 여러 개의 시리즈를 모아놓기 위해 사용, 표 형태로 데이터 출력하고자 사용
인덱스와 컬럼 두 가지 기준에 의하여 표 형태처럼 데이터가 저장 됨

### Pandas 두번째 노드 사용어(클래스) 정리
--------------------------------------------------------------------
**행 삭제**  

|클래스|설명|
|------|----|
|`df.drop('Unnamed: 0', axis=1)`|'Unnamed: 0' 컬럼의 데이터들을 열방향(axis=1)으로 삭제하기|
|`df.drop(columns=['Unnamed: 0', '원산지'])`|리스트 안의 요소에 있는 컬럼이 모두 삭제|  
  
--------------------------------------------------------------------
**결측치 처리**  

|클래스|설명|
|------|----|
|`df.dropna(axis=0)`|결측치가 있는 행 모두 삭제(기본값 axis=0), 결측치가 있는 컬럼 모두 삭제(axis=1)|
|`df.isnull()`|결측치 확인하기 bool 형태의 DF 타입 반환|
|`df.isnull().sum()`|컬럼별 결측치의 개수 확인할 수 있음|
|`df['원산지'].fillna('알수없음')`|'원산지' 컬럼의 결측치 '알수없음'로 채우기|
|`df['칼로리'].fillna(df['칼로리'].min())`|'칼로리' 컬럼의 결측치를 칼로리 컬럼의 최솟값으로 채우기|


--------------------------------------------------------------------
  
**정렬**  

|클래스|설명|
|------|----|
|`df.sort_index(ascending = True)` |(ascending = True) 오름차순 정렬(기본 값), 내림차순 정렬 (ascending = False)|
|`df.sort_values('가격')` |가격 컬럼 기준으로 오름차순 정렬|
|`df.sort_values(['가격', '원산지'], ascending = [False, True])`|복합 적용, 리스트 형으로 묶어주고 출력 할 컬럼 불러오기, 어느 기준으로 어센딩 할 지 각각 리스트로 표현|

--------------------------------------------------------------------
**자료형 변환**  


|클래스|설명|
|------|----|
|`df.reset_index(drop = True)` 	|기존의 인덱스 값이 삭제되고, 리셋이 됨|
|`df.info()` |type 확인 어떤 형식 (DataFrame 등), dtypes, data columns 등 확인 가능|
|`df['호수'] = df['호수'].str.replace('호','')` 	|호수 열 안의 호 삭제, 단어 삭제나 변경|
|`df['호수'] = df['호수'].astype(int)` | 자료형 변환 astype() 괄호 안에 바꿔줄 형태|

--------------------------------------------------------------------
**데이터탐색**  
    

|클래스|설명|
|------|----|
|`df['호수'].mean()`|평균 구하기|
|`df.['호수'].std()`|표준편차 구하기|
|`df.shape`|데이터 프레임 크기 (행, 컬럼) 확인, 괄호 넣지 않고 shape 메소드만|
|`df.describe()`|기초 통계|
|`df.describe(include = 'O')`|기초 통계 (object), O(오)를 입력하면 오브젝트 수가 나옴|
|`df.corr()`|상관관계|
|`df.nunique()`|항목 종류 수|
|`df['호수'].unique()`|항목 종류|
|`df['호수'].value_counts()`|항목별 개수, 유니크보다 보기 편해서 많이 사용|
|`df.count()`|카운트 (컬럼) # 기본 값 axis = 0,    df.count(axis = 1)로 하면 행으로 나옴|
|`df.shape[0]`|행 확인, len(df)으로도 확인 가능|

----------------------------------------------------------------------
**내장 함수**  

  
|클래스|설명|
|------|----|
|`df['가격'].max()`|가격 열의 최대값|
|`df['가격'].min()`|가격 열의 최소값|
|`df['가격'].mean()`|가격 열의 평균값|
|`df['가격'].median()`|가격 열의 중앙값|
|`df['가격'].sum()`|가격 열의 합|
|`df['가격'].std()`|가격 열의 표준편차|
|`df['가격'].var()`|가격 열의 분산|
|`df['가격'].describe()`|가격 열의 백분위수|
|`df['가격'].quantile(0.25)`|하위 25% 값, quantile()괄호 안에 뽑을 값 입력|

--------------------------------------------------------------------
**그룹핑, apply**  


|클래스|설명|
|------|----|
|`cond = df['가격'].quantile(0.25) > df['가격']`|하위 25% 데이터를 전부 뽑아야 하므로 변수 선언 후 수식 입력|
|`df['원산지'].mode()[0]`|최빈값 구하기, .mode()까지만 작성하면 해당 열에 해당하는 유니크 값이 나옴, 리스트 형으로 추출할 값 숫자 입력|
|`df.T`|Transpose, 행과 열 바꿈 (전치행렬)|
|`df.groupby('원산지').mean()`|기준을 적어서 그룹핑을 하고 평균을 출력해야 함, df['원산지'].mean()하면 int형과 str형이 섞여서 에러|
|`df.groupby([ '원산지', '할인율'] )['가격'].mean()`|결과만 보면 Series 형태|
|`pd.DataFrame(df.groupby(['원산지', '할인율'])['가격'].mean())`|데이터프레임으로 다시(?) 묶어줌|
|`df.groupby(['원산지', '할인율'])[['가격']].mean()`|가격 부분에 대괄호를 하나 더 씌워줌으로써 2차원 (DataFrame)을 바로 묶어주기가 가능|
|`df.groupby(['원산지', '할인율']).max()`|원산지와 할인율 기준 최대값|
|`df.groupby(['원산지', '할인율']).max().reset_index()`|1개 인덱스 형태로 리셋|    
  
-----------------------------------------------------------------------  
**apply 함수**    
  
|클래스|설명|
|------|----|
|`df['살찔까요'] = df['칼로리'].apply(cal)`|apply 적용해서 새로운 컬럼 생성, 위의 함수 입력되어서 살찔까요에 해당 조건이 맞으면 Yes or No 출력|
|`df['고민중'] = df['칼로리'].apply(lambda x: '먹지마' if x >= 1300 else '먹어라')`|lambda를 사용해서 바로 간결하게 조건식 적용|
